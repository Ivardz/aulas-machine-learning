# Kernel SVM

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

```python
np.random.seed(1)
X = np.random.randn(200, 2)
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
y = np.where(y, 1, -1)
```

```python
plt.scatter(X[y == 1, 0],
            X[y == 1, 1],
            c='b', marker='x',
            label='1')
plt.scatter(X[y == -1, 0],
            X[y == -1, 1],
            c='r',
            marker='s',
            label='-1')
plt.show()
```

```python
def plot(X, y, kernel, gamma, C):
    svm = SVC(C, kernel, gamma=gamma, random_state=1)
    svm.fit(X, y)
    plot_decision_regions(X, y, svm)
    plt.show()   
```

```python
plot(X, y, kernel='rbf', gamma=0.1, C=1.0)
```

```python
plot(X, y, kernel='rbf', gamma=0.1, C=10.0)
```

```python
plot(X, y, kernel='rbf', gamma=0.1, C=10000.0)
```

```python
plot(X, y, kernel='rbf', gamma=10.0, C=10.0)
```
